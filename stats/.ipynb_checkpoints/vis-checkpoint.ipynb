```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy
```

```python
tr40k=pd.read_csv('40k.csv')
tr40k_3=pd.read_csv('40k_3.csv')
tr40k_4=pd.read_csv('40k_4.csv')
tr20k=pd.read_csv('20k.csv')
tr10k=pd.read_csv('10k.csv')
tr5k=pd.read_csv('5k.csv')
test40k=pd.read_csv('40k_test.csv')
test40k_3=pd.read_csv('40k_test_3.csv')
test40k_4=pd.read_csv('40k_test_4.csv')
test20k=pd.read_csv('20k_test.csv')
test10k=pd.read_csv('10k_test.csv')
test5k=pd.read_csv('5k_test.csv')
print(tr40k_4.head())
test5k.head()
```

```python
#SAME PARAM DIFF BATCHES
#MNB
mnb5k0=test5k[(test5k['classifier']=='mnb') & (test5k['class']==0)].reset_index().drop('index',axis=1)
mnb10k0=test10k[(test10k['classifier']=='mnb') & (test10k['class']==0)].reset_index().drop('index',axis=1)
mnb20k0=test20k[(test20k['classifier']=='mnb') & (test20k['class']==0)].reset_index().drop('index',axis=1)
mnb40k0=test40k[(test40k['classifier']=='mnb') & (test40k['class']==0)].reset_index().drop('index',axis=1)
mnb5k4=test5k[(test5k['classifier']=='mnb') & (test5k['class']==4)].reset_index().drop('index',axis=1)
mnb10k4=test10k[(test10k['classifier']=='mnb') & (test10k['class']==4)].reset_index().drop('index',axis=1)
mnb20k4=test20k[(test20k['classifier']=='mnb') & (test20k['class']==4)].reset_index().drop('index',axis=1)
mnb40k4=test40k[(test40k['classifier']=='mnb') & (test40k['class']==4)].reset_index().drop('index',axis=1)



plt.rcParams["figure.figsize"] = (16,8)
plt.title('MNB accuracy for different batch numbers')
plt.xlabel('Batch No')
plt.ylabel('Accuracy')

plt.plot(mnb5k0['accuracy'],label='5K')
plt.plot(mnb10k0['accuracy'],label='10K')
plt.plot(mnb20k0['accuracy'],label='20K')
plt.plot(mnb40k0['accuracy'],label='40K')
plt.legend()
plt.show()
```

We can see the batch size does not really effect the models accuracy over multiple batches of the test data

```python
#BNB
bnb5k0=test5k[(test5k['classifier']=='bnb') & (test5k['class']==0)].reset_index().drop('index',axis=1)
bnb10k0=test10k[(test10k['classifier']=='bnb') & (test10k['class']==0)].reset_index().drop('index',axis=1)
bnb20k0=test20k[(test20k['classifier']=='bnb') & (test20k['class']==0)].reset_index().drop('index',axis=1)
bnb40k0=test40k[(test40k['classifier']=='bnb') & (test40k['class']==0)].reset_index().drop('index',axis=1)
bnb5k4=test5k[(test5k['classifier']=='bnb') & (test5k['class']==4)].reset_index().drop('index',axis=1)
bnb10k4=test10k[(test10k['classifier']=='bnb') & (test10k['class']==4)].reset_index().drop('index',axis=1)
bnb20k4=test20k[(test20k['classifier']=='bnb') & (test20k['class']==4)].reset_index().drop('index',axis=1)
bnb40k4=test40k[(test40k['classifier']=='bnb') & (test40k['class']==4)].reset_index().drop('index',axis=1)



#plt.xticks(mnbacc5k)
plt.title('BNB accuracy for different batch numbers')
plt.xlabel('Batch No')
plt.ylabel('Accuracy')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(bnb5k0['accuracy'],label='5K')
plt.plot(bnb10k0['accuracy'],label='10K')
plt.plot(bnb20k0['accuracy'],label='20K')
plt.plot(bnb40k0['accuracy'],label='40K')
plt.legend()
plt.show()
```

We again see that batch size does not have much of an effect on the accuracy 

```python
#SGD
sgd5k0=test5k[(test5k['classifier']=='sgd') & (test5k['class']==0)].reset_index().drop('index',axis=1)
sgd10k0=test10k[(test10k['classifier']=='sgd') & (test10k['class']==0)].reset_index().drop('index',axis=1)
sgd20k0=test20k[(test20k['classifier']=='sgd') & (test20k['class']==0)].reset_index().drop('index',axis=1)
sgd40k0=test40k[(test40k['classifier']=='sgd') & (test40k['class']==0)].reset_index().drop('index',axis=1)
sgd5k4=test5k[(test5k['classifier']=='sgd') & (test5k['class']==4)].reset_index().drop('index',axis=1)
sgd10k4=test10k[(test10k['classifier']=='sgd') & (test10k['class']==4)].reset_index().drop('index',axis=1)
sgd20k4=test20k[(test20k['classifier']=='sgd') & (test20k['class']==4)].reset_index().drop('index',axis=1)
sgd40k4=test40k[(test40k['classifier']=='sgd') & (test40k['class']==4)].reset_index().drop('index',axis=1)





plt.title('SGD accuracy for different batch numbers')
plt.xlabel('Batch No')
plt.ylabel('Accuracy')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(sgd5k0['accuracy'],label='5K')
plt.plot(sgd10k0['accuracy'],label='10K')
plt.plot(sgd20k0['accuracy'],label='20K')
plt.plot(sgd40k0['accuracy'],label='40K')
plt.legend()
plt.show()
```

Same here ,it varies from 75% to 77%

```python
#40k training batch size
vecavg=tr40k['vector'].mean()
sgdavg=tr40k['sgd'].mean()
bnbavg=tr40k['bnb'].mean()
mnbavg=tr40k['mnb'].mean()
x_ax=["Vector","mnb","sgd","bnb"]
y_ax=[vecavg,mnbavg,sgdavg,bnbavg]
plt.ylabel('Average execution time per batch (s)')
plt.bar(x_ax,y_ax,color=['blue','red','green','yellow'])
```

We notice that most of the training time for a batch goes into vectorizing the input text. Training the model itself hardly takes any time.

```python
x_ax=["mnb","sgd","bnb"]
y_ax=[mnbavg,sgdavg,bnbavg]
plt.ylabel('Average execution time per batch (s)')
plt.bar(x_ax,y_ax,color=['red','green','yellow'])
```

Clearly Multinomial Naive Bayes has the lowest average execution time.

Now we plot different graphs for the SGD classifier with different hyperparameters for the same batch size of 40k.

```python
#For 40k batch size
#SGD
sgd40k0_3=test40k_3[(test40k_3['classifier']=='sgd') & (test40k_3['class']==0)].reset_index().drop('index',axis=1)
sgd40k4_3=test40k_3[(test40k_3['classifier']=='sgd') & (test40k_3['class']==4)].reset_index().drop('index',axis=1)
sgd40k0_4=test40k_4[(test40k_4['classifier']=='sgd') & (test40k_4['class']==0)].reset_index().drop('index',axis=1)
sgd40k4_4=test40k_4[(test40k_4['classifier']=='sgd') & (test40k_4['class']==4)].reset_index().drop('index',axis=1)



plt.title('SGD accuracy for different HyperparameterS')
plt.xlabel('Batch No')
plt.ylabel('Accuracy')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(sgd40k0['accuracy'],label='loss=log')
plt.plot(sgd40k0_3['accuracy'],label='loss=squared_hinge,penalty=l1')
plt.plot(sgd40k0_4['accuracy'],label='penalty=elasticnet')

plt.legend()
plt.show()
```

```python

plt.title('SGD precision(Class 0) for different HyperparameterS')
plt.xlabel('Batch No')
plt.ylabel('Precision')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(sgd40k0['precision'],label='loss=log')
plt.plot(sgd40k0_3['precision'],label='loss=squared_hinge,penalty=l1')
plt.plot(sgd40k0_4['precision'],label='penalty=elasticnet')

plt.legend()
plt.show()

plt.title('SGD precision(Class=4) for different HyperparameterS')
plt.xlabel('Batch No')
plt.ylabel('Precision')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(sgd40k4['precision'],label='loss=log')
plt.plot(sgd40k4_3['precision'],label='loss=squared_hinge,penalty=l1')
plt.plot(sgd40k4_4['precision'],label='penalty=elasticnet')

plt.legend()
plt.show()

```

```python
plt.title('SGD f1-score(Class 0) for different HyperparameterS')
plt.xlabel('Batch No')
plt.ylabel('F1-score')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(sgd40k0['f1-score'],label='loss=log')
plt.plot(sgd40k0_3['f1-score'],label='loss=squared_hinge,penalty=l1')
plt.plot(sgd40k0_4['f1-score'],label='penalty=elasticnet')

plt.legend()
plt.show()

plt.title('SGD f1-score(Class=4) for different HyperparameterS')
plt.xlabel('Batch No')
plt.ylabel('F1-score')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(sgd40k4['f1-score'],label='loss=log')
plt.plot(sgd40k4_3['f1-score'],label='loss=squared_hinge,penalty=l1')
plt.plot(sgd40k4_4['f1-score'],label='penalty=elasticnet')

plt.legend()
plt.show()
```

We see that squared_hinged loss and l1 penalty performs poorly in our models across different metrics. We notice that default parameters with penalty as elasticnet slighly outperforms log loss in our test.

Now we compare the different MNB models made using the different hyperparameters just like we did with SGD.
