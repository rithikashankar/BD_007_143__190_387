```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy
```

```python
tr40k=pd.read_csv('40k.csv')
tr40k_3=pd.read_csv('40k_3.csv')
tr40k_4=pd.read_csv('40k_4.csv')
tr20k=pd.read_csv('20k.csv')
tr10k=pd.read_csv('10k.csv')
tr5k=pd.read_csv('5k.csv')
test40k=pd.read_csv('40k_test.csv')
test40k_3=pd.read_csv('40k_test_3.csv')
test40k_4=pd.read_csv('40k_test_4.csv')
test20k=pd.read_csv('20k_test.csv')
test10k=pd.read_csv('10k_test.csv')
test5k=pd.read_csv('5k_test.csv')
print(tr40k_4.head())
test5k.head()
```

```python
#SAME PARAM DIFF BATCHES
#MNB
mnb5k=test5k.loc[test5k['classifier']=='mnb'].reset_index().drop('index',axis=1)
mnb10k=test10k.loc[test10k['classifier']=='mnb'].reset_index().drop('index',axis=1)
mnb20k=test20k.loc[test20k['classifier']=='mnb'].reset_index().drop('index',axis=1)
mnb40k=test40k.loc[test40k['classifier']=='mnb'].reset_index().drop('index',axis=1)



#plt.xticks(mnbacc5k)
plt.title('MNB accuracy for different batch numbers')
plt.xlabel('Batch No')
plt.ylabel('Accuracy')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(mnb5k['accuracy'],label='5K')
plt.plot(mnb10k['accuracy'],label='10K')
plt.plot(mnb20k['accuracy'],label='20K')
plt.plot(mnb40k['accuracy'],label='40K')
plt.legend()
plt.show()
```

```python
#BNB
bnb5k=test5k.loc[test5k['classifier']=='bnb'].reset_index().drop('index',axis=1)
bnb10k=test10k.loc[test10k['classifier']=='bnb'].reset_index().drop('index',axis=1)
bnb20k=test20k.loc[test20k['classifier']=='bnb'].reset_index().drop('index',axis=1)
bnb40k=test40k.loc[test40k['classifier']=='bnb'].reset_index().drop('index',axis=1)



#plt.xticks(mnbacc5k)
plt.title('BNB accuracy for different batch numbers')
plt.xlabel('Batch No')
plt.ylabel('Accuracy')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(bnb5k['accuracy'],label='5K')
plt.plot(bnb10k['accuracy'],label='10K')
plt.plot(bnb20k['accuracy'],label='20K')
plt.plot(bnb40k['accuracy'],label='40K')
plt.legend()
plt.show()
```

```python
#SGD
sgd5k=test5k.loc[test5k['classifier']=='sgd'].reset_index().drop('index',axis=1)
sgd10k=test10k.loc[test10k['classifier']=='sgd'].reset_index().drop('index',axis=1)
sgd20k=test20k.loc[test20k['classifier']=='sgd'].reset_index().drop('index',axis=1)
sgd40k=test40k.loc[test40k['classifier']=='sgd'].reset_index().drop('index',axis=1)



plt.title('SGD accuracy for different batch numbers')
plt.xlabel('Batch No')
plt.ylabel('Accuracy')
plt.rcParams["figure.figsize"] = (16,8)
plt.plot(sgd5k['accuracy'],label='5K')
plt.plot(sgd10k['accuracy'],label='10K')
plt.plot(sgd20k['accuracy'],label='20K')
plt.plot(sgd40k['accuracy'],label='40K')
plt.legend()
plt.show()
```
